```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Bob', 'Alice', 'Alex', 'Bob', 'Alice'],
    'Age': [23, 56, 12, 30, 35],
    'Salary': [50000, 60000, 40000, 50000, 60000]
}

df = pd.DataFrame(data)

# Ensure that 'Salary' column is numeric
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

# Drop rows with NaN values, just in case
df = df.dropna()

# Group by 'Salary' and calculate the mean for other columns
result = df.groupby('Salary').mean(numeric_only=True)

print(result)
```

```python
import pandas as pd

# Create two DataFrames
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

# Concatenate the DataFrames vertically (along rows)
result = pd.concat([df1, df2], ignore_index=True)

# Print df1, df2, and the concatenated result
print("df1:")
print(df1)
print("--------")
print("df2:")
print(df2)
print("--------")
print("result:")
print(result)
print("--------------")

# Perform groupby on the 'B' column of the concatenated DataFrame and calculate the mean of each group
grouped_result = result.groupby('B').mean()

# Print the grouped result
print(grouped_result)
```

```python
import pandas as pd

# Step 1: Create sample data
data = {
    'Date': ['2021-01-01', '2021-01-03'],  # notice 2021-01-02 is missing
    'Price': [100, 102]
}
df = pd.DataFrame(data)

# Step 2: Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Step 3: Set 'Date' as index (resample only works on index)
df.set_index('Date', inplace=True)

# Step 4: Upsample to daily frequency and forward fill
df_upsampled = df.resample('D').ffill()

# Step 5: Show results
print("Upsampled with Forward Fill:")
print(df_upsampled)
print("\nOriginal DF:")
print(df)
```

```python
data = {
    'Date': ['2021-01-01','2021-01-02', '2021-01-03'],  # notice 2021-01-02 is missing
    'Price': [100,102,103]
 }
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df['Shifted']=df['Price'].shift(1)
print(df)
print("------------")
df['Shifted Backward']=df['Price'].shift(-1)
print(df)
```

```python
data = {
    'Date': pd.date_range(start='2021-01-01', periods=10, freq='D'),
    'Price': [100, 102, 103, 105, 107, 110, 108, 107, 111, 115]

}
df=pd.DataFrame(data)
print(df)
df.set_index('Date',inplace=True)

monthly=df.resample('ME').mean()
print(monthly)
weekly=df.resample('W').mean()
print(weekly)
```

```python
data = {
    'Date': pd.date_range(start='2021-01-01', periods=5, freq='D'),
    'Price': [100, 102, 104, 108, 110]
}
df=pd.DataFrame(data)
df.set_index('Date',inplace=True)
df['Expanding_Mean']=df['Price'].expanding().mean()
print(df)
df.plot()
```

```python
data = {'Price': [5, 10, 15, 20, 25, 30]}
df = pd.DataFrame(data)
def myrange(x):
    return (x.max()-x.min())/2
df['Rolling range']=df['Price'].rolling(window=3).apply(myrange)
print(df)
df.plot()
```

```python
import pandas as pd

# Sample time series data with missing values
data = {'Date': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05'],
        'Price': [100, None, 105, None, 110]}
df = pd.DataFrame(data)

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Show the DataFrame with missing values
print("Original DataFrame:")
print(df)

# Forward Fill
df_ffill = df.fillna(method='ffill')

# Backward Fill
df_bfill = df.fillna(method='bfill')

# Interpolation
df_interp = df.interpolate(method='linear')

# Drop missing values
df_dropna = df.dropna()

# Display the results
print("\nForward Fill:")
print(df_ffill)

print("\nBackward Fill:")
print(df_bfill)

print("\nInterpolated:")
print(df_interp)

print("\nDropped Missing Values:")
print(df_dropna)
```

```python
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [1, 2, 4], 'Age': [25, 30, 22]})

# Merge on 'ID' (similar to SQL JOIN)
merged_df = pd.merge(df1, df2, on='ID', how='inner')  # 'inner' means only common IDs
print(merged_df)

# how='inner': Returns only rows with matching keys in both DataFrames.

# how='left': Returns all rows from the left DataFrame, and matching rows from the right DataFrame.

# how='right': Returns all rows from the right DataFrame, and matching rows from the left DataFrame.

# how='outer': Returns all rows from both DataFrames, with NaN for missing values.

merged_df.plot()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {'Age': [22, 25, 27, 30, 35, 40, 45, 50, 55, 60],
        'Salary': [2500, 2800, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500]}

df = pd.DataFrame(data)

# Descriptive Statistics
print(df.describe())

# Checking for missing values
print(df.isnull().sum())

# Histogram of Age
sns.histplot(df['Age'], kde=True)
plt.show()

# Boxplot for Salary
sns.boxplot(x=df['Salary'])
plt.show()
plt.plot(df['Age'], df['Salary'])
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
sns.histplot(df['Age'],kde=True)
plt.show()
sns.scatterplot(x='Age', y='Salary', data=df)
plt.show()
sns.pairplot(df)
plt.show()
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()



```
