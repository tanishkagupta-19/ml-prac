# ðŸŒŸ Basic Setup

```python
# Cell 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create some sample data
df = pd.DataFrame({
    'Age': [22, 25, 47, 52, 46, 56, 48, 35, 24, 50],
    'Salary': [25000, 27000, 52000, 58000, 49000, 60000, 52000, 42000, 26000, 54000],
    'Department': ['HR', 'HR', 'Engineering', 'Engineering', 'Engineering', 
                   'Management', 'Management', 'HR', 'HR', 'Management']
})
df.head()
```

# ðŸ“Š 1. Histogram â€” Distribution of Single Variable
ðŸ”µ Purpose: See how data is spread (like age distribution).

```python
# Cell 2: Histogram of Age
plt.figure(figsize=(3,3))
sns.histplot(df['Age'],kde=True, color='pink')
plt.title('Distribution of AGE')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid('True')
plt.show()
```

# ðŸ“Š 2. Boxplot â€” Detect Outliers and Spread
ðŸ”µ Purpose: Find outliers and understand the data spread.

```python
plt.figure(figsize=(4,4))
sns.boxplot(x=df['Salary'],color='pink')
plt.title("Boxplot of Salary")
plt.xlabel('Salary')
plt.grid(True)
plt.show()
```

# ðŸ“Š 3. Bar Plot â€” Categorical Comparison
ðŸ”µ Purpose: Compare mean values across categories.

```python
plt.figure(figsize=(4,4))
sns.barplot(x='Department',y='Salary',data=df,palette='husl')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.grid(True)
plt.show()
```

# ðŸ“Š 4. Scatter Plot â€” Relationship Between Two Variables
ðŸ”µ Purpose: See correlations/relationships between two continuous variables.

```python
plt.figure(figsize=(4,4))
sns.scatterplot(x='Age',y='Salary',data=df,hue='Department',palette='husl')
plt.title("Age vs Salary")
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

# ðŸ“Š 5. Line Plot â€” Trend Over Time
ðŸ”µ Purpose: See trends over sequences (time or ordered data).

```python
# Cell 6: Line plot example (Age vs Salary just for demo)
plt.figure(figsize=(4,4))
sns.lineplot(x='Age', y='Salary', data=df, marker='o', color='pink')
plt.title('Line Plot of Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

# ðŸ“Š 6. Heatmap â€” Correlation Between Features



```python
plt.figure(figsize=(4,4))
sns.heatmap(df.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation heatmap')
plt.show()
# annot makes the number inside the colums appear
# corr forms correalation between the a and y
```

# ðŸ“Š 7. Pairplot â€” Plot Every Pair Against Each Other

```python
plt.figure(figsize=(8,6))
sns.pairplot(df,hue='Department', palette='husl')
plt.suptitle('Pairwise Plots',y=1.02)
plt.show()
```

# 4ï¸âƒ£ Violin Plot (boxplot + distribution)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
df = pd.DataFrame({
    'Department': ['HR', 'HR', 'HR', 'Engineering', 'Engineering', 'Engineering', 
                   'Management', 'Management', 'Management', 'HR', 'Engineering', 'Management'],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 
               'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Salary': [28000, 30000, 32000, 50000, 52000, 48000, 
               70000, 72000, 69000, 29000, 51000, 70500]
})

# Create violin plot with color by Gender
plt.figure(figsize=(4,4))
sns.violinplot(x='Department', y='Salary', hue='Gender', data=df, palette='Set2', split=True)
plt.title('Salary Distribution by Department and Gender', fontsize=14)
plt.ylabel('Salary', fontsize=12)
plt.xlabel('Department', fontsize=12)
plt.legend(title='Gender')
plt.grid(True)
plt.tight_layout()
plt.show()
```
