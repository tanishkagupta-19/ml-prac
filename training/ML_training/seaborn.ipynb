```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset('tips')
```

```python
# df1=sns.load_dataset('titanic')
# df1.head()
```

```python
print('Sample Dataset(Tips):')
df.head()
```

```python
print('Data Information:')
df.info()
```

```python
print('Statistical Summary:')
df.describe()
```

```python
print("Scatter Plot: Total Bill vs Tip:")
plt.figure(figsize=(7,5))
sns.scatterplot(data=df,x='total_bill',y='tip')
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()
```

```python
print("Scatter Plot: Total Bill vs Tip:")
plt.figure(figsize=(7,5))
sns.scatterplot(data=df,x='total_bill',y='tip',hue='sex',size='smoker')
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()
```

```python
print("Bar Plot: Footfall by Day:")
plt.figure(figsize=(7,5))
sns.countplot(data=df,x='day',hue='sex')
plt.title("Bar Plot: Footfall by Day")
plt.show()
```

```python
print("Bar Plot: Footfall by Day:")
plt.figure(figsize=(7,5))
sns.countplot(data=df,x='day',hue='time')
plt.title("Bar Plot: Footfall by Day")
plt.show()
```

```python
print("Box plot for total bill per day:")
plt.figure(figsize=(7,5))
sns.boxplot(data=df,x='day',y='total_bill',hue='sex')
plt.title("Box plot for total bill per day")
plt.show()
```

```python
print("Swarm plot for total bill per day:")
plt.figure(figsize=(7,5))   
sns.swarmplot(data=df,x='day',y='total_bill',hue='sex')
plt.title("Swarm plot for total bill per day")
plt.show()
```

```python
# only for continuous(numerical) variables
print("Heatmap for correlation:")
plt.figure(figsize=(4,4))
sns.heatmap(df.select_dtypes(include='number').corr(),annot=True,cmap='coolwarm')
plt.title("Heatmap for correlation")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.pairplot(df,hue='sex')
plt.suptitle("Pairplot: Relationship among features",y=1.02)
plt.show()
```
