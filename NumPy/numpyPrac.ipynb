```python
import numpy as np
```

CREATING ARRAY

```python
a=np.array([1,2,3])
print(a)
b=np.array([[1,2,3],[4,5,6]])
print(b)
```

SPECIAL ARRAYS

```python
a=np.zeros((3,4))  # array of zeros
print(a)
b=np.ones((2,3))  #array of ones
print(b)
c=np.eye(3)   #identity matrix
print(c)
d=np.arange(0,20,3)  #(from,to,step)
print(d)
e=np.linspace(0,5,3) #(from,to,number of random values)
print(e)
```

ARRAY ATTRIBUTES

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a.shape) #(rows,cols)
print(a.ndim) #(total number of dimensions)
print(a.size) #no of elements
print(a.dtype) #data type of elements
```

ARRAY INDEXING AND SLICING

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr[0, 1])  # element at 1st row, 2nd column => 2
print(arr[:, 1])     # all rows, 2nd column => [2, 5]
print(arr[1, :])     # 2nd row => [4, 5, 6]
```

ARRAY OPERTIONS

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)         # [5, 7, 9]
print(a * b)         # [4, 10, 18]
print(a ** 2 )       # [1, 4, 9]

```

BROADCASTING

```python
a = np.array([[1], [2], [3]])
b = np.array([10, 20, 30])
print(a + b)
```

MULTIPLICATION

```python
a=np.array([[1,2],[4,5]])
b=np.array([[6,7],[2,3]])
print(np.dot(a,b))
```

TRANSPOSE

```python
print(a.T)
```

INVERSE

```python
a=np.array([[1,2],[4,5]])
b=np.array([[6,7],[2,3]])
print(np.linalg.inv(a))
print(np.linalg.inv(b))
```

DETERMINANT

```python
arr=np.array([[56,43],[39,41]])
print(np.linalg.det(arr))
```

AGGREGATION FUNCTIONS

```python
arr=np.array([[1,2,3],[4,5,6],[6,7,8]])
print(np.sum(arr)) #total sum
print(np.mean(arr)) #mean
print(np.max(arr, axis=0)) #max of each col
print(np.min(arr, axis=1)) # min of each row
print(np.std(arr)) # standard deviation
```

MASKING AND FILTERING

```python
arr=np.array([2,3,4,6,7,8])
mask=arr>3
print(arr[mask])
```

RESHAPING & STACKING

```python
s="Random Module"
print(s.upper())
```

```python
arr = np.arange(6).reshape(2, 3)   # reshape to 2 rows, 3 cols
print(arr)
a = np.array([1, 2])
b = np.array([3, 4])
print(np.vstack([a, b])  )  # vertical stack
print(np.hstack([a, b]))    # horizontal stack
```

*RANDOM MODULE

```python
np.random.seed(69) # produces the same result (pics same 3 random numbers everytime)
print(np.random.rand(4)) 
print(np.random.rand(3,3)) # pics random number in matrix of size 2x3 from range [0,1)
print(np.random.randn(2,3)) #centered around 0(can be negative)
print(np.random.randint(0,10,(2,3))) # random nums from 0 to 10 of size 2x3
```

USEFUL UTILITIES

```python
arr=np.array([3,4,1,2,6])
print(np.unique([1,2,3,4,5,5])) # behaves like set
print(np.where(arr>3,1,0)) #np.where(condition , replace by x if true, replace by y if false)
print(np.clip(arr,2,4)) # np.clip(arr,min,max) values below min will be replaced by min and values above max will be replaced by max 
```

STANDARDIZE TO MAKE MEAN=0 DEVIATION=1

```python
data = np.array([12, 15, 20, 21, 18])
mean = np.mean(data)
std = np.std(data)

standardized = (data - mean) / std
print(standardized)
print(np.mean(standardized))
print(np.std(standardized))

```
