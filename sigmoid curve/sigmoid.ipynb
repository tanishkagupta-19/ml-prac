```python
import numpy as np
import matplotlib.pyplot as plt

# Input values (already between 0 and 1)
x = np.array([0.13, 0.21, 0.34, 0.57, 0.72, 0.89])

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Apply sigmoid
y = sigmoid(x)

# Plot
plt.plot(x, y, marker='o', linestyle='-', color='green')
plt.title("Sigmoid Curve")
plt.xlabel("Input (x)")
plt.ylabel("Sigmoid Output (Ïƒ(x))")
plt.grid(True)
plt.show()
```
